## Exercise 4

** Use plt.subplots(nrows=1, ncols=2)**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. Played around with the linewidth and style**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y, color='blue', lw='2.50', ls='--')
axes[1].plot(x,z, color='red', lw='3.00')
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots()**

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize= (12,2))

axes[0].plot(x,y, color='blue', lw='4.00')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z, color='red', lw='2.50', ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')

fig
```
